```python
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm

import pickle
```

```python
#a random dataframe with 4 yeast strains (assume you isolated these strains for beer production)

samples = pd.read_csv('../data/all_data_13.csv',index_col=0)
samples = samples.sample(n=4)
filename_sample = '../data/predictions/samples.csv'
samples.to_csv(filename_sample, index=True)
proteins_measured = samples.iloc[:,0:725] #from 0 to 725 are ORFs (proteins), from 726 to 738 are the metabolites
proteins_measured.index = ['Strain1', 'Strain2', 'Strain3', 'Strain4']
filename_example = '../data/predictions/example.csv'
proteins_measured.to_csv(filename_example, index=True)
```
